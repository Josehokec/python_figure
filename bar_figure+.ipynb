```python
"""
柱状图画图代码参考示例
By 南京大学计算机学院刘世喆
"""
import matplotlib.pyplot as plt
import numpy as np

titile_font = {'family':'Arial', 'weight': 'bold', 'size': 15}
legende_font = {'family':'Arial', 'weight': 'bold', 'size': 13}
colors = ['#6be8a8', '#6bace8', '#e8a86b', '#e86b6d', '#e86bac']
        
x = [1, 2, 3, 4]
x = np.array(x)
index_ls = ['JOB', 'Crimes', 'NASDAQ', 'SD10M']
values1 = [66, 538, 504, 504]
values2 = [343, 1818, 1488, 1288]
values3 = [252, 1627, 1143, 1543]
values4 = [100, 420, 320, 320]


plt.figure(figsize= (5, 3))
plt.xticks(x, index_ls, fontweight='bold')

     
# 设置x和y轴数值显示字体大小
plt.xticks(fontsize=14.5) 
# 设置y轴数值显示字体大小为
plt.yticks(fontsize=13)     

# x轴和y轴截断
plt.xlim(0.5, 4.5)
plt.ylim(50, 9000)

total_width= 0.85
method_num = 4
width = total_width / method_num
x = x - (total_width - width) / 2

plt.bar(x, values1, label = 'Base', color = colors[0], width = width, edgecolor='black', linewidth=1)
plt.bar(x + width, values2, label = 'Base+O1', color = colors[1],width=width, edgecolor='black', linewidth=1)
plt.bar(x + 2 * width, values3, label = 'Base+O2', color =  colors[2], width=width, edgecolor='black', linewidth=1)
plt.bar(x + 3 * width, values4, label = 'Base+O1+O2', color = colors[3], width=width, edgecolor='black', linewidth=1)

plt.yscale('log')

"""
handlelength‌：图例标记的长度‌
‌handletextpad‌：图例标记与标签之间的间距‌
‌labelspacing‌：图例中各标签之间的垂直间距‌
ncol‌：图例列数，默认值为1，表示图例将按列排列
bbox_to_anchor‌：图例的锚点位置，用于调整图例的位置
"""
plt.legend(ncol = 2, prop=legende_font, loc='upper left', labelspacing = 0.3)
plt.ylabel('Average Latency (ms)', titile_font)

# plt.grid()
plt.grid(axis='y', color='gray', linestyle='--')

plt.tight_layout()
plt.savefig(fname='optimization.pdf')    # 先保存再展示
plt.show()
```
